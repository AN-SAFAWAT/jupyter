```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
#popularity count plo

plt.figure(figsize=(12,15))
sns.countplot(y='popularity',data=file)
```

```python
file['duration']=file['duration_ms']/1000
file.drop('duration_ms',axis=1,inplace=True)
file[['duration','popularity']].sort_values('duration',ascending=False)
```

```python
sns.scatterplot(x='popularity', y='duration',data=file)
```

```python
file.columns
```

```python
file['release_date']=pd.to_datetime(file['release_date'])
file.release_date
file['year']=file.release_date.dt.year
file['month']=file.release_date.dt.month
file['month_name']=file.release_date.dt.month_name

```

```python


#average popularity in term of year
mean_pop=file[['popularity','year']].groupby('year',as_index=False).mean()
mean_dura=file[['duration','year']].groupby('year',as_index=False).mean()

plt.subplot(1,2,1)
sns.scatterplot(x='year',y='popularity',data=mean_pop,markers='*')
plt.subplot(1,2,2)
sns.scatterplot(x='year',y='duration',data=mean_dura,markers='*')
```

we see that popularity of the songsbincreasing day by day
#duration of songs increased from 1990 to 2000 but after that it started decreasing again

```python
group_by=file[['popularity','year','month']].groupby(['year','month'],as_index=False).sum()
group_by1=group_by.sort_values('popularity',ascending=False).head(30)
group_by1

```

```python
comparing_2020=file[file['year']==2020][['energy','loudness','year']]
comparing_2000=file[file['year']==2020][['energy','loudness','year']]
comparing_1990=file[file['year']==1990][['energy','loudness','year']]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
sns.scatterplot(x='energy',y='loudness',data=comparing_2020,ax=axes[0,0])

sns.scatterplot(x='energy',y='loudness',data=comparing_2000,ax=axes[0,1])

sns.scatterplot(x='energy',y='loudness',data=comparing_1990,ax=axes[1,0])

sns.scatterplot(x='energy',y='loudness',data=comparing_1990,ax=axes[1,1])
```
